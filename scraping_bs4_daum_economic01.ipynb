```python
path = 'http://media.daum.net/economic/'

req = requests.get(path)
```

```python
req.content
```

```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(req.content, 'html.parser')
```

```python
type(soup)
```

```python
result = soup.select('div > strong.tit_thumb > a[href].link_txt')
type(result)
```

```python
result[0].text
```

```python
tag = result[0]
```

```python
tag['href']
```

[
    [title, link].

```python
contents = [];
for tag in result:
    # print(tag.text.strip(), tag['href'].strip())
    title = tag.text.strip()
    link = tag['href'].strip()
    content = [title, link]
    contents.append(content)
    
len(contents)
```

```python
contents
```

```python
import pandas as pd
```

```python
pd = pd.DataFrame(contents, columns=['title','link'])

pd
```

```python
pd.to_excel('./saves/economic01.xls', index=False)
```
